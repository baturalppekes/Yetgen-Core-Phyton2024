# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
def vki_hesapla(kilo, boy):
    # VKİ hesaplama formülü
    return kilo / (boy ** 2)

def vki_siniflandir(vki):
    if vki < 18.5:
        return "Zayıf"
    elif 18.5 <= vki < 25:
        return "Normal kilolu"
    elif 25 <= vki < 30:
        return "Fazla kilolu"
    elif 30 <= vki < 40:
        return "Obez"
    else:
        return "İleri derecede obez (morbid obez)"

def veri_girisi_yap():
    giris_sayisi = int(input("Kaç giriş yapacaksınız? "))

    with open("kullanici_bilgileri.txt", "a", encoding="utf-8") as dosya:
        for i in range(giris_sayisi):
            print(f"{i+1}. kullanıcı bilgilerini girin:")
            isim_soyisim = input("İsim Soyisim: ")
            yas = int(input("Yaş: "))
            boy = float(input("Boy (metre cinsinden, örn: 1.75): "))
            kilo = float(input("Kilo: "))

            vki = vki_hesapla(kilo, boy)
            sinif = vki_siniflandir(vki)

            # Bilgileri dosyaya yaz
            dosya.write(f"{isim_soyisim}, {yas}, {boy}, {kilo}, {vki:.2f}, {sinif}\n")

            # Bilgileri ekrana yazdır
            print(f"İsim Soyisim: {isim_soyisim}, Yaş: {yas}, Boy: {boy}, Kilo: {kilo}, VKİ: {vki:.2f}, Sınıf: {sinif}\n")

# Veri giriş fonksiyonunu çağırarak işlemi başlat
veri_girisi_yap()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import datetime

class YetBar:
    def __init__(self):
        self.menu = {
            "Banana": 1.0,
            "Strawberry": 1.5,
            "Blueberry": 2.0,
            "Mango": 2.5,
            "Protein Powder": 3.0,
            "Spinach": 1.0,
            "Almond Milk": 1.5,
            "Peanut Butter": 2.0,
        }
    
    def menu_goster(self):
        print("Menüdeki ürünler ve fiyatları:")
        for urun, fiyat in self.menu.items():
            print(f"{urun}: ${fiyat:.2f}")
    
    def satin_al(self):
        toplam_fiyat = 0
        secilen_urunler = []
        
        while True:
            urun = input("Lütfen ürünü seçin (çıkmak için 'q' ya basın): ").title()
            if urun == 'Q':
                break
            elif urun in self.menu:
                miktar = float(input(f"Kaç tane {urun} almak istiyorsunuz? "))
                fiyat = self.menu[urun] * miktar
                toplam_fiyat += fiyat
                secilen_urunler.append((urun, miktar, fiyat))
            else:
                print("Bu ürün menüde yok, lütfen geçerli bir ürün seçin.")
        
        if toplam_fiyat > 0:
            tarih = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            with open("satin_alma_log.txt", "a", encoding="utf-8") as dosya:
                dosya.write(f"Tarih: {tarih}, Toplam Fiyat: ${toplam_fiyat:.2f}, Ürünler: {secilen_urunler}\n")
            print(f"Toplam Fiyat: ${toplam_fiyat:.2f}")
            print("Satın alma işlemi başarıyla kaydedildi.")
        else:
            print("Satın alma işlemi iptal edildi.")

def main():
    yetbar = YetBar()
    
    while True:
        print("YetBar'a hoş geldiniz! Lütfen yapmak istediğiniz işlemi seçin:")
        print("1. Menüyü görüntüleme")
        print("2. Satın alma")
        print("3. Çıkış")
        
        secim = input("Seçiminiz (1/2/3): ")
        
        if secim == '1':
            yetbar.menu_goster()
        elif secim == '2':
            yetbar.satin_al()
        elif secim == '3':
            print("Çıkış yapılıyor...")
            break
        else:
            print("Geçersiz seçim, lütfen tekrar deneyin.")

if __name__ == "__main__":
    main()
```
