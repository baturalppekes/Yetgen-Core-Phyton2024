# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random

class Kumanda:
    def __init__(self):
        self.tv_acik = False
        self.ses_seviyesi = 10
        self.kanallar = []
        self.tv_bilgileri = {
            "Marka": "Unknown",
            "Model": "Unknown",
            "Yıl": "Unknown"
        }
    
    def tv_ac(self):
        if not self.tv_acik:
            self.tv_acik = True
            print("Televizyon açıldı.")
        else:
            print("Televizyon zaten açık.")

    def tv_kapat(self):
        if self.tv_acik:
            self.tv_acik = False
            print("Televizyon kapatıldı.")
        else:
            print("Televizyon zaten kapalı.")

    def ses_ayarları(self):
        if self.tv_acik:
            print(f"Mevcut ses seviyesi: {self.ses_seviyesi}")
            ses_seviyesi = int(input("Yeni ses seviyesini girin (0-100): "))
            if 0 <= ses_seviyesi <= 100:
                self.ses_seviyesi = ses_seviyesi
                print(f"Ses seviyesi ayarlandı: {self.ses_seviyesi}")
            else:
                print("Geçersiz ses seviyesi. Ses seviyesi 0 ile 100 arasında olmalıdır.")
        else:
            print("Televizyon kapalı. Ses ayarları yapılamaz.")

    def kanal_ekle(self):
        if self.tv_acik:
            kanal = input("Eklemek istediğiniz kanalın adını girin: ")
            if kanal not in self.kanallar:
                self.kanallar.append(kanal)
                print(f"{kanal} kanalı eklendi.")
            else:
                print("Bu kanal zaten mevcut.")
        else:
            print("Televizyon kapalı. Kanal eklenemez.")
    
    def kanal_sayisini_ogrenme(self):
        if self.tv_acik:
            print(f"Toplam kanal sayısı: {len(self.kanallar)}")
        else:
            print("Televizyon kapalı. Kanal sayısı öğrenilemez.")
    
    def rastgele_kanala_gecme(self):
        if self.tv_acik:
            if self.kanallar:
                kanal = random.choice(self.kanallar)
                print(f"Rastgele seçilen kanal: {kanal}")
            else:
                print("Listede kanallar mevcut değil.")
        else:
            print("Televizyon kapalı. Kanal değiştirilemez.")
    
    def televizyon_bilgileri(self):
        print("Televizyon Bilgileri:")
        for bilgi, deger in self.tv_bilgileri.items():
            print(f"{bilgi}: {deger}")

    def kumanda(self):
        while True:
            print("\nYapmak istediğiniz işlemi seçin:")
            print("1. Tv Aç")
            print("2. Tv Kapat")
            print("3. Ses Ayarları")
            print("4. Kanal Ekle")
            print("5. Kanal Sayısını Öğrenme")
            print("6. Rastgele Kanala Geçme")
            print("7. Televizyon Bilgileri")
            print("q. Çıkış")

            secim = input("Seçiminizi yapın: ")
            
            if secim == '1':
                self.tv_ac()
            elif secim == '2':
                self.tv_kapat()
            elif secim == '3':
                self.ses_ayarları()
            elif secim == '4':
                self.kanal_ekle()
            elif secim == '5':
                self.kanal_sayisini_ogrenme()
            elif secim == '6':
                self.rastgele_kanala_gecme()
            elif secim == '7':
                self.televizyon_bilgileri()
            elif secim == 'q':
                print("Kumanda kapatılıyor...")
                break
            else:
                print("Geçersiz seçim. Lütfen tekrar deneyin.")

# Kumanda sınıfının bir örneğini oluşturun ve kumanda fonksiyonunu çağırın
kumanda = Kumanda()
kumanda.kumanda()
```
